```python
import pandas as pd
```

```python
df = pd.read_csv('legends.txt', header=None, names=['Champion'])
dfAll = df.sort_values(by=['Champion'], ascending=True)
dfAll.reset_index(drop=True, inplace=True)
dfAll.head()
dfAll.to_csv('champList.csv', index=False)
```

```python
df = pd.read_csv("top.csv", header=None, names=['Champion'])
dfTop = df.sort_values(by=['Champion'], ascending=True)
dfTop.reset_index(drop=True, inplace=True)
dfTop['Role'] = 'Top'
dfTop.head()
```

```python
df = pd.read_csv("jg.csv", header=None, names=['Champion'])
dfJg = df.sort_values(by=['Champion'], ascending=True)
dfJg.reset_index(drop=True, inplace=True)
dfJg['Role'] = 'Jungle'
dfJg.head()
```

```python
df = pd.read_csv('mid.csv', header=None, names=['Champion'])
dfMid = df.sort_values(by=['Champion'], ascending=True)
dfMid.reset_index(drop=True, inplace=True)
dfMid['Role'] = 'Mid'
dfMid.head()
```

```python
df = pd.read_csv('bot.csv', header=None, names=['Champion'])
dfBot = df.sort_values(by=['Champion'], ascending=True)
dfBot.reset_index(drop=True, inplace=True)
dfBot['Role'] = 'Bot'
dfBot.head()
```

```python
df = pd.read_csv('supp.csv', header=None, names=['Champion'])
dfSupp = df.sort_values(by=['Champion'], ascending=True)
dfSupp.reset_index(drop=True, inplace=True)
dfSupp['Role'] = 'Support'
dfSupp.head()
```

```python
dfRoles = pd.concat([dfTop, dfJg, dfMid, dfBot, dfSupp])
dfRoles.head()
```

```python
champion_roles = {}
for champion in dfAll['Champion']:
    champion_roles[champion] = []

for df_role in [dfTop, dfJg, dfMid, dfBot, dfSupp]:
    for _, row in df_role.iterrows():
        champion = row['Champion']
        if champion in champion_roles:
            champion_roles[champion].append(row['Role'])
            
# create a list of dictionaries, where each dictionary represents a row in the output df
data = []
for champion, roles in champion_roles.items():
    row = {'Champion': champion}
    for role in ['Top', 'Jungle', 'Mid', 'Bot', 'Support']:
        row[role] = (role in roles)
    data.append(row)
    
# create the output df
dfOutput = pd.DataFrame(data)
dfOutput = dfOutput[['Champion', 'Top', 'Jungle', 'Mid', 'Bot', 'Support']]
dfOutput.head()

dfOutput.to_csv('roles.csv', index=False)
```
